# データ理解

## Instrument（計測手段）

```python
data_dict['Instrument'].unique()
```

- **Identifier（識別子）**：参加者を識別するためのIDを含むデータ
- **Demographics（人口統計）**：参加者に関する基本的な人口統計データ（年齢、性別、登録シーズンなど）が含まれる
- **Children's Global Assessment Scale（CGAS: 子どもの全般的評価尺度）**：子どもの全般的な機能を評価するための尺度で、参加者の機能レベルを季節ごとに評価するためのデータが含まれる
  - CGAS-CGAS_Scoreはよく分からない
- **Physical Measures（身体測定）**：身体的な特徴や状態を測定するデータです。身長、体重、BMI（ボディマス指数）など、身体の成長や健康状態に関する基本的なデータ
- **FitnessGram Vitals and Treadmill（フィットネスグラム・バイタルとトレッドミル）**：FitnessGramプログラムに基づくフィットネス評価の一部で、バイタルサイン（心拍数、血圧など）やトレッドミルテストの結果
- **FitnessGram Child（フィットネスグラム・チャイルド）**：子どもの身体的なフィットネスレベルを測定するための標準化されたテスト（柔軟性、筋力、持久力などの評価結果）
- **Bio-electric Impedance Analysis（生体インピーダンス解析）**：体内の電流の流れを基に体脂肪率、筋肉量、水分量などを推定する測定方法
- **Physical Activity Questionnaire (Adolescents)（身体活動質問票（青少年））**：青少年を対象とした身体活動に関する質問票
- **Physical Activity Questionnaire (Children)（身体活動質問票（子ども））**：子どもを対象とした身体活動に関する質問票
- **Parent-Child Internet Addiction Test（親子インターネット依存テスト）**：親子を対象にインターネット依存の度合いを測定するためのテスト
- **Sleep Disturbance Scale（睡眠障害尺度）**：睡眠の質や障害の程度を評価するための尺度
- **Internet Use（インターネット使用）**：インターネットの利用に関する情報

### FitnessGram Child
- **FGC-Season**：測定が行われた季節
- **FGC-FGC_CU**：腹筋の回数
- **FGC-FGC_CU_Zone**：カールアップのフィットネスゾーン（0=改善が必要、1=健康的なフィットネスゾーン。）
- **FGC-FGC_GSND**： 握力（非利き手）
- **FGC-FGC_GSND_Zone**：握力（非利き手）のフィットネスゾーン（1=弱い、2=標準、3=強い）
- **FGC-FGC_GSD**： 握力（利き手）
- **FGC-FGC_GSD_Zone**：握力（利き手）のフィットネスゾーン（1=弱い、2=標準、3=強い）
- **FGC-FGC_PU**：腕立ての回数
- **FGC-FGC_PU_Zone**：プッシュアップのフィットネスゾーン（0=改善が必要、1=健康的なフィットネスゾーン。）
- **FGC-FGC_SRL**：座位体前屈（左側）の総値。
- **FGC-FGC_SRL_Zone**：座位体前屈（左側）のフィットネスゾーン。（0=改善が必要、1=健康的なフィットネスゾーン。）
- **FGC-FGC_SRR**：座位体前屈（右側）の総値。
- **FGC-FGC_SRR_Zone**：座位体前屈（右側）のフィットネスゾーン。（0=改善が必要、1=健康的なフィットネスゾーン。）
- **FGC-FGC_TL**：トランクリフト（体幹リフト）の総値
- **FGC-FGC_TL_Zone**：トランクリフトのフィットネスゾーン。（0=改善が必要、1=健康的なフィットネスゾーン。）
